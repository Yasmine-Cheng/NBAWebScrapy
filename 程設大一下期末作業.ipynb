```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup
import re      #正規運算式的模組

driver=webdriver.Chrome()
driver.get('https://nba.udn.com/nba/index?gr=www')
driver.find_element_by_id("redirectNBA").click()
time.sleep(3)
driver.fullscreen_window()
driver.find_element_by_partial_link_text('Teams').click()
time.sleep(3)

name=[]
position=[]
height=[]
weight=[]


htmltext=driver.page_source
soup=BeautifulSoup(htmltext,"lxml")
    
x=1
for times in range(len(soup.select('.team__list'))-2):
    
    htmltext=driver.page_source
    soup=BeautifulSoup(htmltext,"lxml")
    for nameinfo in soup.select(".nba-player-index__name"):
        name.append(nameinfo.text)
    for otherinfo in soup.select(".nba-player-index__details"):
        text=otherinfo.text
        
        if 'Forward-Center' in text:   
            position.append('Forward-Center')
        elif 'Guard-Forward' in text:   
            position.append('Guard-Forward')
        elif 'Guard' in text:  
            position.append('Guard')
        elif 'Forward' in text: 
            position.append('Forward')
        else:
            position.append('Center')
        clear=re.findall(r'[0-9]+|[a-z]+',text)
        clearone=(int(clear[-6])*30.48)+(int(clear[-4])*2.54) 
        height.append(clearone)
        clearonee=int(clear[-2])*0.45359237
        weight.append(clearonee)
        
    x=x+1
    time.sleep(3)
    driver.find_elements_by_class_name("team__list a")[x].click()
    time.sleep(3)
    
htmltext=driver.page_source
soup=BeautifulSoup(htmltext,"lxml")
for info in soup.select(".nba-player-index__name"):
    name.append(info.text)
for otherinfo in soup.select(".nba-player-index__details"):
    text=otherinfo.text
        
    if 'Forward-Center' in text: 
        position.append('Forward-Center')
    elif 'Guard-Forward' in text: 
        position.append('Guard-Forward')
    elif 'Guard' in text:  
        position.append('Guard')
    elif 'Forward' in text:   
        position.append('Forward')
    else:
        position.append('Center')
    clear=re.findall(r'[0-9]+|[a-z]+',text)     #把數字跟英文分開
    clearone=(int(clear[-6])*30.48)+(int(clear[-4])*2.54) 
    height.append(clearone)
    clearonee=int(clear[-2])*0.45359237
    weight.append(clearonee)
print(name,position,weight,height)
```

```python
import pandas as pd

df=pd.DataFrame({'Name':name,'Position':position,'Weight(kg)':weight,'Height(cm)':height})
df.to_csv('nba.csv',index=False,encoding='utf_8_sig')
df
```

```python
df['BMI']=df['Weight(kg)']/df['Height(cm)']/df['Height(cm)']*100*100
```

```python
df=df.round(2)
df
```

```python
from bokeh .plotting import figure,output_file,show
from bokeh.models import CategoricalColorMapper
import matplotlib.pyplot as plt   

map=CategoricalColorMapper(factors=['Guard-Forward','Guard','Center','Forward','Forward-Center'],palette=['red','orange','yellow','green','blue'])

p=figure(title='位置與身材的關係圖(體重&身高)')

p.circle(x='Weight(kg)',y='Height(cm)',source=df,size=10,color={'field':'Position','transform':map})

show(p)
```
